A Vector is typically used to store and search Text, Video, Sound or other nummerical information.

In this LAB we will use it to store nummerical values (vectors) that describes measured sound signals from different loudspeakers.

Oracle23ai new datatype Vector is the latest new technology that was added to Oracle23ai in early summer of 2024.

Traditionally Relational Databases are  used for searching and finding exact values.

Vectors on the other hand is very usefull to find similar records - in this example finding speakers that have the most identical sound.

The datatype of the vectors must be identical, and can be:

INT8 (8-bit integers)
FLOAT32 (32-bit IEEE floating-point numbers)
FLOAT64 (64-bit IEEE floating-point numbers)
BINARY (packed UINT8 bytes where each dimension is a single bit)

In this lab we will be using float32, but int would probably have been more optimal.

An example of a Binary Vector:

CREATE TABLE vectab(id NUMBER, data VECTOR(16, BINARY));
INSERT INTO vectab VALUES (1, '[201, 15]');

Vectors are internally stored as Securefile BLOBs and most popular embedding model vector sizes are between 1.5KB and 12KB in size. 
You can use the following formula to determine the size of your vectors on disk: 
number of vectors * number of dimensions * size of your vector dimension type (for example, a FLOAT32 is equivalent to BINARY_FLOAT and is 4 bytes in size).

The alternative to use Vectors in the Oracle database, is to use a specialised Vector database.
However with Vectors inside Oracle - it is possible to search using combinations of exact search on "structured" columns and Vectors.
